```python
def cat_num(data):
    categorical = []
    numerical = []
    
    for i in data.columns:
        if data[i].dtypes == 'O':
            categorical.append(i)
        else:
            numerical.append(i)
            
    return categorical, numerical

categorical, numerical = cat_num(data)
```

```python
from tabulate import tabulate
print(tabulate({"Categorical":categorical,
               "Numerical":numerical},headers=["Ctaegorical","Numerical"],tablefmt="fancy_grid"))
```

```python
def treat_null(data, categorical):
    for i in data.columns:
        if i in categorical:
            data[i].fillna(data[i].mode()[0], inplace=True)
        else:
            data[i].fillna(data[i].mode()[0], inplace=True)

treat_null(data, categorical)
```
